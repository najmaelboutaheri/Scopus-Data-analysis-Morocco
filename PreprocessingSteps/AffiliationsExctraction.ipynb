````python
# Filter rows that contain "Morocco" to target Moroccan affiliations
moroccan_affiliations = data_affiliation_author[data_affiliation_author['affiliation'].str.contains('Morocco', case=False)]
````

**2. Exctract  affiliations based on 'University' word including this word in diffrent languages using the following regular expression:**

````python
# Pre-compile the regex pattern for efficiency
pattern = re.compile(r"([^,]*?)\s*(university|université|universidad)[\s\-]*(.*?)(?=,|$)", re.IGNORECASE)

# Function to extract the full university name
def extract_university_name(affiliation):
    match = pattern.search(affiliation)
    if match:
        # Concatenate the full university name by combining matched groups
        university_name = f"{match.group(1).strip()} {match.group(2).strip()} {match.group(3).strip()}"
        return university_name.strip()  # Remove any leading or trailing spaces
    else:
        return "Unknown"

# Apply the extraction function to the affiliation column
moroccan_affiliations['university_name'] = moroccan_affiliations['affiliation'].apply(extract_university_name)
````
