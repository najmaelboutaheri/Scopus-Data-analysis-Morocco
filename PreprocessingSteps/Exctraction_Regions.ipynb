```python
# Ensure all values in 'affiliation' are strings
affiliation_combined_dataset['affiliation'] = affiliation_combined_dataset['affiliation'].astype(str)
```

```python
import re

# Task 1: Extract Moroccan Cities
def extract_city(affiliation):
    # Ensure the affiliation is a string
    if isinstance(affiliation, str) and affiliation.endswith("Morocco"):
        for city in moroccan_cities:
            # Ensure city is a string before comparison
            if isinstance(city, str) and city in affiliation:
               print(affiliation)
               return city
    return None

affiliation_combined_dataset['City'] = affiliation_combined_dataset['affiliation'].apply(extract_city)

# Task 2: Extract Countries
def extract_country(affiliation):
    for country in countries:
        if country in affiliation:
            return country
    return None

affiliation_combined_dataset['Country'] = affiliation_combined_dataset['affiliation'].apply(extract_country)

# Save the results
affiliation_combined_dataset.to_csv("affiliation_combined_dataset-with-regions.csv", index=False)
```
